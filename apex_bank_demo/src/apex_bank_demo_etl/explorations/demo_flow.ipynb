## PII Masking

Views can have custom masking functions applied per-column

Benefits:
* Dynamic Data Masking: Masks data at query time, ensuring sensitive information is protected without altering the underlying data.
* Granular Control: Allows for fine-grained control over which users or groups can view unmasked data.
* Centralized Governance: Managed within Unity Catalog, promoting consistent data governance policies.
* Compliance: Helps meet data privacy and compliance requirements by masking sensitive data fields.
* Flexibility: Masking functions can be customized to implement various masking techniques (redaction, encryption, tokenization, etc.)

```python
%sql
-- Query the masked view (what data scientists see)
SELECT 
    transaction_id,
    account_id,  -- Shows as XXXX-XXXX-XXXX-1234
    transaction_timestamp,
    amount,
    merchant_category_desc,
    merchant_name,
    card_present_flag,
    is_fraud
FROM apex_bank_demo.analytics.transactions_silver_masked
LIMIT 10;
```
